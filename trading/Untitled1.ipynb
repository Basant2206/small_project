```python
import yfinance as yf
from datetime import datetime, timedelta

# Define the date range
end_date = datetime.today()
start_date = end_date - timedelta(days=5*365)  # 5 years ago

# Try alternate ticker if ^NSEI fails
nifty50_index = '^NSEI'  # First attempt
alternative_ticker = 'NIFTY50'  # Potential alternative

try:
    print(f"Fetching data for ticker: {nifty50_index}...")
    nifty_data = yf.Ticker(nifty50_index).history(start=start_date, end=end_date)
    
    if nifty_data.empty:
        print(f"No data found for {nifty50_index}. Trying alternative ticker...")
        nifty_data = yf.Ticker(alternative_ticker).history(start=start_date, end=end_date)
    
    if nifty_data.empty:
        print("Still no data found. Please check available tickers or use a different source.")
    else:
        print("Nifty 50 Index Data (Last 5 Years):")
        print(nifty_data.tail())
        nifty_data.to_csv('nifty50_data.csv', index=True)
        print("Data has been saved to 'nifty50_data.csv'.")
except Exception as e:
    print(f"Failed to fetch data: {e}")
```

```python
import pandas as pd
import requests
from datetime import datetime

# Define the URL for Nifty 50 historical data
url = "https://www1.nseindia.com/products/content/indices/historical_index_data.htm"

# Define the date range
end_date = datetime.today().strftime('%d-%m-%Y')
start_date = (datetime.today() - pd.DateOffset(years=1)).strftime('%d-%m-%Y')

# Parameters for the request
params = {
    'indexType': 'NIFTY 50',
    'fromDate': start_date,
    'toDate': end_date
}

# Send a GET request to download the data
try:
    print("Fetching Nifty 50 Index Data...")
    response = requests.get(url, params=params)
    response.raise_for_status()  # Raise an error for bad status codes

    # Load the data into a DataFrame
    nifty_data = pd.read_csv(pd.compat.StringIO(response.text))
    if nifty_data.empty:
        print("No data found for Nifty 50. Please check the URL and parameters.")
    else:
        print("Nifty 50 Index Data (Last 5 Years):")
        print(nifty_data.tail())  # Display the last few rows
except Exception as e:
    print(f"Failed to fetch Nifty 50 data: {e}")
```
