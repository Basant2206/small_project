```python
from datetime import datetime, timedelta
import yfinance as yf
import pandas as pd
import numpy as np

# List of company
nifty50_tickers = ['^NSEI',
    'RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'ICICIBANK.NS', 'INFY.NS', 'HINDUNILVR.NS', 
    'ITC.NS', 'KOTAKBANK.NS', 'SBIN.NS', 'BHARTIARTL.NS', 'LT.NS', 'BAJFINANCE.NS', 
    'ASIANPAINT.NS', 'HCLTECH.NS', 'MARUTI.NS', 'TITAN.NS', 'AXISBANK.NS', 'SUNPHARMA.NS', 
    'NTPC.NS', 'ONGC.NS', 'NESTLEIND.NS', 'ULTRACEMCO.NS', 'WIPRO.NS', 'ADANIPORTS.NS', 
    'POWERGRID.NS', 'M&M.NS', 'BAJAJ-AUTO.NS', 'GRASIM.NS', 'TECHM.NS', 'JSWSTEEL.NS', 
    'TATASTEEL.NS', 'DRREDDY.NS', 'BRITANNIA.NS', 'EICHERMOT.NS', 'SHREECEM.NS', 
    'INDUSINDBK.NS', 'DIVISLAB.NS', 'COALINDIA.NS', 'UPL.NS', 'BPCL.NS', 'IOC.NS', 
    'CIPLA.NS', 'HEROMOTOCO.NS', 'GAIL.NS', 'SBILIFE.NS', 'HDFCLIFE.NS', 'TATAMOTORS.NS', 
    'VEDL.NS'
]

# fetching data for 'year = 5'
def fetch_data(company, year):
    end_date = datetime.today()
    start_date = end_date - timedelta(days=year*365)
    
    # Fetch historical data for Nifty 50
    df = yf.download(company, start=start_date, end=end_date)
    return (df[(  'High', company)]+df[(  'Low', company)])/2
# for each company fetching 1 data per day
# which is mean of low and high of that perticular day

data= pd.DataFrame()

# here is main code for fetch data
for i in nifty50_tickers:
    data[i] = fetch_data(i,10)
```

```python
# initialising available amount for perticular company and 
amount = {'^NSEI':500000}
invest={'^NSEI':500000}
for i in data.columns[1:]:
    amount[i] = 500000/len(data.columns[1:])
    invest[i] = 0
    
```

```python
# code for buy and sall
for date in data.index:
    for col in data.columns:
        if data.loc[date,col] == np.nan:
            invest[col]=0
        

# code to buy and sell
```
