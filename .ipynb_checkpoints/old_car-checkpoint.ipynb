```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['selling_price']), df['selling_price'], test_size=0.2, random_state=2)
```

```python
X_train.head()
```

```python
oe=OrdinalEncoder(categories=[['Test Drive Car','Fourth & Above Owner','Third Owner','Second Owner','First Owner']])
oe.set_output(transform='pandas') # as by default it give numpy array, to get in pandas dataframe
X_train_owner=oe.fit_transform(X_train.loc[:,['owner']])
X_test_owner=oe.fit_transform(X_test.loc[:,['owner']])
```

```python
X_train_owner
```

```python
ohe = OneHotEncoder(sparse_output=False,handle_unknown='ignore',min_frequency=100)
ohe.set_output(transform = 'pandas')
X_train_brandfuel=ohe.fit_transform(X_train[['brand','fuel']])
X_test_brandfuel=ohe.transform(X_test[['brand','fuel']])
```

```python
X_train_brandfuel
```

```python
ss=StandardScaler()
ss.set_output(transform = 'pandas')
km_driven_train=ss.fit_transform(X_train[['km_driven']])
km_driven_test=ss.transform(X_test[['km_driven']])
```

```python
final_train=pd.concat([X_train_owner,X_train_brandfuel,km_driven_train],axis=1)
final_test=pd.concat([X_test_owner,X_test_brandfuel,km_driven_test],axis=1)
```

# Smart Way

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(
    [
        ("ordinal", OrdinalEncoder(categories=[['Test Drive Car','Fourth & Above Owner','Third Owner','Second Owner','First Owner']]),['owner']),
        ("onehot", OneHotEncoder(sparse_output=False,handle_unknown='ignore',min_frequency=100), ['brand','fuel']),
        ("scaler", StandardScaler(),['km_driven'])
    ],
    remainder='passthrough'
)
```

```python
transformer
```

```python
X_train_transformer = transformer.fit_transform(X_train)
X_test_transformer = transformer.transform(X_test)

X_train_transformed=pd.DataFrame(X_train_transformer, columns = transformer.get_feature_names_out())
X_test_transformed=pd.DataFrame(X_test_transformer, columns = transformer.get_feature_names_out())
```
